```python
import pytesseract
from pdf2image import convert_from_path
from transformers import pipeline
from google.colab import files
import os
```

```python
# Upload a PDF file
uploaded = files.upload()

# Get the filename
pdf_path = list(uploaded.keys())[0]
print(f"✅ Uploaded File: {pdf_path}")

```

```python
# Convert PDF pages to images
images = convert_from_path(pdf_path)

# Extract text from each image using Tesseract OCR
extracted_text = ""
for i, img in enumerate(images):
    extracted_text += pytesseract.image_to_string(img, lang="eng") + f"\n\n--- Page {i+1} ---\n\n"

# Save the raw extracted text
with open("extracted_text.txt", "w") as file:
    file.write(extracted_text)

print("✅ Text extraction completed! Saved as 'extracted_text.txt'")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load T5 model and tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Preprocess text: add a task prefix
input_text = "fix: " + extracted_text[:512]  # Limit to 512 tokens

# Tokenize input
inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

# Generate enhanced text
outputs = model.generate(**inputs, max_length=512)

# Decode output
improved_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Save enhanced text
with open("enhanced_text.txt", "w") as file:
    file.write(improved_text)

print("✅ Enhanced text saved as 'enhanced_text.txt'")
```

```python
import pytesseract
from pdf2image import convert_from_path
from transformers import BartForConditionalGeneration, BartTokenizer, T5ForConditionalGeneration, T5Tokenizer
import torch
!pip install evaluate

```

```python
# Path to the uploaded PDF
pdf_path = '/content/frdA290120A2141312_FirstRanker.com (1) (1).pdf'  # Replace with the correct path of the PDF

# Convert PDF to images
images = convert_from_path(pdf_path, 300)

# Display the first image (optional)
images[0].show()
```

```python
# Extract text from the images
texts = []
for img in images:
    text = pytesseract.image_to_string(img)
    texts.append(text)

# Combine all extracted text from the pages
extracted_text = "\n".join(texts)
print(extracted_text[:1000])  # Print the first 1000 characters for verification
```

```python
# Load pre-trained BART model and tokenizer
bart_model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
bart_tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")

# Summarize with BART
inputs = bart_tokenizer(extracted_text, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = bart_model.generate(inputs['input_ids'], num_beams=4, max_length=150, early_stopping=True)

# Decode the generated summary
bart_summary = bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("BART Summary:", bart_summary)
```

```python
# Load pre-trained T5 model and tokenizer
t5_model = T5ForConditionalGeneration.from_pretrained("t5-large")
t5_tokenizer = T5Tokenizer.from_pretrained("t5-large")

# Prepare input for T5
t5_input = "summarize: " + extracted_text
inputs_t5 = t5_tokenizer(t5_input, return_tensors="pt", max_length=1024, truncation=True)

# Summarize with T5
summary_ids_t5 = t5_model.generate(inputs_t5['input_ids'], num_beams=4, max_length=150, early_stopping=True)

# Decode the generated summary
t5_summary = t5_tokenizer.decode(summary_ids_t5[0], skip_special_tokens=True)
print("T5 Summary:", t5_summary)
```

```python
from google.colab import files
import pytesseract
from pdf2image import convert_from_path
import cv2
import numpy as np

# Upload PDF file
uploaded = files.upload()

# Extract the file name from the uploaded file
pdf_path = next(iter(uploaded))  # Assuming only one file is uploaded

# Function to preprocess the image before OCR
def preprocess_image(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)

    # Apply thresholding (adjust parameters as needed)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

    # Optional: Apply noise removal or other preprocessing techniques
    # thresh = cv2.medianBlur(thresh, 3)

    return thresh

# Function to extract text from PDF using OCR with preprocessing
def extract_text_from_pdf(pdf_path):
    images = convert_from_path(pdf_path)

    extracted_text = ""
    for image in images:
        preprocessed_image = preprocess_image(image)
        text = pytesseract.image_to_string(preprocessed_image, config='--psm 6')  # PSM 6 is suitable for uniform text blocks
        extracted_text += text + "\n"

    return extracted_text

# Extract and print the text from the uploaded PDF
extracted_text = extract_text_from_pdf(pdf_path)
print("Extracted Text from PDF:")
print(extracted_text)
```

```python
import re

def clean_text(text):
    # Remove extra newlines
    cleaned_text = re.sub(r'\n+', '\n', text)
    # Remove extra spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    return cleaned_text

# Apply the cleaning function to the extracted text
cleaned_text = clean_text(extracted_text)
print("Cleaned Text:")
print(cleaned_text)
```
